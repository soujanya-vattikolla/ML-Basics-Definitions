Build decision tree model to predict survival based on certain parameters

```python
import pandas as pd
```

```python
titantic_df = pd.read_csv('titantic.csv')
titantic_df.head()
```

```python
X = titantic_df[['Pclass', 'Sex', 'Age', 'Fare']]
y = titantic_df[['Survived']]
```

```python
X.head()
```

```python
y.head()
```

```python
# converting to numerical values for column 'Sex'

from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
X['sex_numerical'] = le_sex.fit_transform(X['Sex'])
X
```

```python
# dropping sex as it is not required

X = X.drop(['Sex'],axis='columns')
X
```

```python
y.head()
```

```python
X.Age.isnull().sum()
```

We can observe that age has 177 NaN values

```python
# Data Preprocessing: filling the NA value with median value of a column (Age)
import math
age = math.floor(X['Age'].median())
age
```

```python
# assinging the value
X['Age'] = X['Age'].fillna(age)
X.Age.isnull().sum()
```

```python
# import tree
from sklearn import tree
decision_model = tree.DecisionTreeClassifier()
```

```python
# training the model
decision_model.fit(X,y)
```

```python
# accuracy of the model
decision_model.score(X,y)
```

We have got an accuracy of 97%.
